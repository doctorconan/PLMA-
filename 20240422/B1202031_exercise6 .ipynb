```python
i=1
while i<=9:
  j=1
  while j<=9:
    p=i*j
    if p<10:
      print(f"{j} × {i} = {p}" ,end='   ')
    else:
      print(f"{j} × {i} = {p}" ,end='  ')

    j+=1
  print()
  i+=1
```
