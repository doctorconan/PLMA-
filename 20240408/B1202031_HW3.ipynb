```python
a=int(input("輸入年份(西元):"))
if a<=1582:
  if a%4==0:
    print('閏年')
  else:
    print('平年')
else:
  if a%400==0:
    print('閏年')
  elif a%100==0:
    print('平年')
  elif a%4==0:
    print('閏年')
  else:
    print('平年')
```
