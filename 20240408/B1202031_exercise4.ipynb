```python
a=int(input("輸入身高(單位:公分):"))
b=int(input("輸入體重(單位:公斤):"))

a/=100
BMI=b/(a*a)
print("BMI:",BMI)

if BMI<18.5:
  print("體位:過輕")
elif 18.5<BMI<24 or BMI==18.5:
  print("體位:適中")
elif 24<BMI<27 or BMI==24:
  print("體位:過重")
else:
  print("體位:肥胖")
```

```python
a=int(input("請輸入第一個數字:"))
b=input("請輸入運算符號:")
c=int(input("請輸入第二個數字:"))

if b=='+':
  print(a+c)
elif b=='-':
  print(a-c)
elif b=='*':
  print(a*c)
elif b=='/':
  print(a/c)
else:
  print()
```
